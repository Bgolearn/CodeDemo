Assign a function to compare the optimization efficiency across different utility functions.



```python
import Bgolearn.BGOsampling as BGOS
import numpy as np
import pandas as pd

# Define the true function
def function(X):
    X = np.array(X)
    Y = 0.013 * X**4 - 0.25 * X**3 + 1.61 * X**2 - 4.1 * X + 8
    return Y

# defind the search space
vs = np.arange(0, 10.01, 0.01)

# Initialize Bgolearn
Bgolearn = BGOS.Bgolearn()

# Pass parameters to the function
Mymodel = Bgolearn.test(Ture_fun=function, Def_Domain=vs)

# Compare efficiency using different methods
Mymodel.Trail()
```
